```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("1364-img0102.png")
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
plt.figure(figsize=(20,10))
imagem_roigray = cv2.cvtColor(imagem[220:380, 520:680], cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_roigray, cmap = "gray")
plt.title("Corpo inteiro")
```

PARA PODER FUNDIR AS IMAGENS É NECESSARIOS ELAS POSSUIREM O MESMO TIPO E QUANTIDADE DE CANAIS!!!!

```python
imagem_grayforrgb = cv2.cvtColor(imagem_roigray, cv2.COLOR_GRAY2RGB)
imagem_rgb[220:380, 520:680] = imagem_grayforrgb
plt.figure(figsize=(20,10))
plt.imshow(imagem_rgb)
plt.title("RESULTADO")
```
